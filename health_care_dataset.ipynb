visualization

```python
import matplotlib.pyplot as plt
ds["Medical Condition"].value_counts().plot(kind="bar",color="pink")

plt.title("Disease Distribution",fontsize=15,color="red")
plt.xlabel("Medical Condition",fontsize=10)
plt.ylabel("Count",fontsize=10)

plt.show()
```

```python
ds["Gender"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")
plt.title("Gender Distribution")
plt.legend(["Male", "Female"])

plt.show()
```

```python
plt.hist(ds["Billing Amount"], bins=50)

plt.title("Treatment Cost Distribution")

plt.show()
```
